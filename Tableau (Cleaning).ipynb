```python
import pandas as pd
```

```python
file_path='co2_emission.csv'
df=pd.read_csv("C:/Users/Chaitali/OneDrive/Desktop/co2_emission.csv")
```

```python
df.rename(columns={
    'Entity':'Country',
    'Code':'Country_Code',
    'Year':'Year',
    df.columns[3]:'CO2_Emissions_Tonnes'
},inplace=True)
```

```python
df=df[['Country','Country_Code','Year','CO2_Emissions_Tonnes']]
```

```python
df['CO2_Emissions_Tonnes'] = pd.to_numeric(df['CO2_Emissions_Tonnes'], errors='coerce')  
df.dropna(subset=['CO2_Emissions_Tonnes'], inplace=True) 
```

```python
df.to_csv('cleaned_co2_emissions.csv', index=False, sep=',')
print("✅ Dataset cleaned and saved as 'cleaned_co2_emissions.csv'")
```
